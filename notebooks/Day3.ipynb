8. Missing Value Analysis

Identify the number of missing values present in each column. This helps determine which features require cleaning or imputation.

Task Covered: Log missing-value counts per column.

```python
missing_events = df.isnull().sum()

missing_events
```

9. Missing Value Percentage

Calculate the percentage of missing values in every column to evaluate the severity of incomplete data.

```python
missing_percent = (df.isnull().mean()*100).round(2)

pd.DataFrame({
    "Missing %": missing_percent
})
```

10. Duplicate Record Analysis (Event_ID)

Check for duplicate Event_ID values in the Landslide Events dataset to ensure each landslide event has a unique identifier.

Task Covered: Identify duplicates using Event_ID.

```python
event_duplicates = df.duplicated(subset=['Event_ID'])

print("Duplicate Event_IDs:", event_duplicates.sum())
```

11. Duplicate Record Analysis (Response_ID)

Check for duplicate Response_ID values in the Emergency Response dataset to verify that each response record is unique.

Task Covered: Identify duplicates using Response_ID.

```python
response_duplicates = df.duplicated(subset=['Response_ID'])

print("Duplicate Response_IDs:", response_duplicates.sum())
```

12. Complete Duplicate Rows

Identify fully duplicated rows across each dataset to detect redundant records that may need removal during the cleaning phase.

```python
print(df.duplicated().sum())
```

13. Unique Values Analysis

Count the number of unique values in each column to better understand feature diversity and detect potential anomalies.
